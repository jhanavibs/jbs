```python
#Reading of csv file
bbq = pd.read_csv('bbq_summer.csv')
print(bbq)
```

```python
#Dropping of BundleID
bbq1=bbq.drop(['bundleID'],axis=1)
```

```python
#Dataset post drop operation
bbq1
```

```python
#checking for null values
bbq1.isnull().values.any()
```

```python
#Sorting of the dataset in a descending manner based on avg_rating
bbq1.sort_values(by='avg_rating', ascending=False)
```

```python
#Obtaining dummies
bbq2 = pd.get_dummies(bbq1, drop_first=True, columns=['starter', 'maindishI', 'maindishII','side','dessert'])
```

```python
bbq2.head()
```

```python
#viewing of columns post generation of dummies
bbq2.columns
```

```python
#assigning input and outut variables
X = bbq2[['starter_Fried Chicken Tenders',
       'starter_Jumbo Shrimp Cocktail', 'starter_Sticky Chicken Tenders',
       'maindishI_BBQ Chicken', 'maindishI_Buffalo Chicken Macaroni Stacker',
       'maindishI_Pulled Pork', 'maindishII_Lobster Roll',
       'maindishII_Sausage', 'maindishII_Steak Sampler', 'side_Mac and Cheese',
       'side_Potato Salad', 'dessert_Blondie Brownie']]
y = bbq2['avg_rating']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
#Intercept value
regressor.intercept_
```

```python
#Coefficients for the dataset
coef_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coef_df
```

```python
regressor.coef_
```

Conjoinint analysis has been performed here for the choosing of food items provided by the vendor for summer grill menu. Recommendations for barbeque options are based on the consumer rating. The coefficients generated here provides better understanding of food preference and the strength of opinions of customers.On analysing the coefficients the model suggests:        
              

**Starter**- Sticky Chicken Tenders , Coefficient value - 0.339275 , Price- $2.9 

For Main Dish I, the model suggests Buffalo Chicken Macaroni Stacker of coefficient value 0.536821 and Price of $6.1 to be the best. But Pulled Pork is rated equally good by the customer rating. So we will be going with Pulled pork option for main dish I.

**Main Dish I**- Pulled Pork, Coefficient value - 0.529312, Price - $5.7

For Main Dish II, the model suggests Lobster Roll seems to be one of the best seller at Lobster Land.

**Main Dish II**- Lobster Roll, Coefficient value - 1.048096, Price - $5.2

**Side dish**- Mac and Cheese, Coefficient Value -0.575125 ,Price- $0.45

**Dessert**- Apple Pie a la Mode, with a positive Coefficient value, Price- $1.1

Above mentioned are the recommendations based on the analysis to Lobster land management.    
Total price - $15.35



As the Lobster Land management has decided to charge a flat $15 fee for all visitors who enter the Barbeque Tent at the Portland Calling festival the food recommendations are evident to attract more customers thereby maximising the revenue and foot fall at the festival.
